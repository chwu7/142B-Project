```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load master index once labels.py has been run
df = pd.read_parquet('../data/processed/master_index.parquet')
print(df.shape)
df.head()
```

```python
LABEL_NAMES = {0: "down", 1: "flat", 2: "up"}

# --- Class distribution ---
counts = df["label"].value_counts().sort_index()
counts.index = [LABEL_NAMES[i] for i in counts.index]

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Bar chart
axes[0].bar(counts.index, counts.values, color=["#d62728", "#aec7e8", "#2ca02c"])
axes[0].set_title("Label Counts")
axes[0].set_ylabel("Count")
for i, v in enumerate(counts.values):
    axes[0].text(i, v + 1, str(v), ha="center", fontweight="bold")

# Pie chart
axes[1].pie(counts.values, labels=counts.index, autopct="%1.1f%%",
            colors=["#d62728", "#aec7e8", "#2ca02c"], startangle=90)
axes[1].set_title("Label Proportions")

plt.suptitle("Class Distribution", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

print(counts.to_string())
print(f"\nTotal samples: {len(df)}")
print(f"Imbalance ratio (max/min): {counts.max() / counts.min():.2f}x")
```
